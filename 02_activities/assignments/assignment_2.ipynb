
-   Create 1 new example that demonstrates you understand the problem. Trace/walkthrough 1 example that your partner made and explain it.


```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

# My Example:
tree = TreeNode(1,
    TreeNode(2,
        TreeNode(4),
        TreeNode(2)  # Duplicate node
    ),
    TreeNode(3)
)
#[[1,2,4],[1,2,2],[1,3]]


# Partners example
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)
# [[1,2,4],[1,2,5],[1,3,6],[1,3,7]]
```


-   Copy the solution your partner wrote. 


```python
def bt_path(root: TreeNode, path = [], result = []) -> list[list[int]]:
    # If there is no value for the node, exit
    if not root:
        return

    # Append current node to path
    path.append(root.val) 
    
    # Add completed path if the node is a leaf
    if not root.left and not root.right: 
        result.append(list(path))
    
    # Explore left first
    bt_path(root.left, path, result) 
    bt_path(root.right, path, result)
    path.pop() # go back if exploration is done

result = []
bt_path(root, [], result)
result

result_2 = []
bt_path(tree, [], result_2)
result_2
```


-   Explain why their solution works in your own words.


```python
# This solution works by recursively calling itself on the left and right node. It keeps track of two empty arrays called path and result. For each node it adds the value into the path array, it then checks
# if it is a leaf node it appends the whole path array into the result array. It then calls itself recursively on the left and right nodes to continue traversing and it pops the value of path so it can remove the most recent node and traverse back up the tree
# without needing to go back and start directly from the root.
```


-   Explain the problem’s time and space complexity in your own words.


```python
# The time complexity is O(N**2) as O(N) nodes are visited and then in a worst case time complexity of a skewed tree it is copying an array of size N into result  so it will be O(N*N) or O(N**2)
# The same goes for Space complexity as it has to keep each visted node into memory and then keep track of an array of size N worst case scenario therefore the solution is O(N**2)
```
